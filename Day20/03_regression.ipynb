### 构建数据集

```python
from torch.utils.data import Dataset  # 从PyTorch中导入Dataset类

# 定义一个自定义数据集类
class HousingDataset(Dataset):
    def __init__(self, mode='train'):
        """
        初始化方法，用来加载数据并进行预处理。
        
        Args:
            mode (str, optional): 数据集的模式，'train'、'valid' 或 'test'，用于选择不同的数据。
        """
        # dataset_maps 是一个预定义的字典，其中包含了训练集、验证集、测试集的数据。
        # 根据传入的 mode ('train'、'valid' 或 'test') 获取对应的数据（特征和标签）
        self.x, self.y = dataset_maps[mode]  # x: 特征，y: 标签，假设它们是NumPy ndarray类型
        
        # 对特征数据 x 进行缩放处理，通常是标准化（例如：均值为0，标准差为1）
        # scaler 是一个预先定义的标准化器（如StandardScaler或MinMaxScaler）
        # 将NumPy数组转换为PyTorch的Tensor，并转换为float类型
        self.x = torch.from_numpy(scaler.transform(self.x)).float()  # 将特征转换为Tensor
        
        # 将标签 y 转换为Tensor，并确保它是一个列向量（多行1列）
        # 使用reshape(-1, 1)确保每个标签是一个单列向量
        self.y = torch.from_numpy(self.y).float().reshape(-1, 1)  # 将标签转换为Tensor，并进行维度调整
    
    def __len__(self):
        """
        返回数据集的长度，也就是数据集中样本的数量。
        
        Returns:
            int: 数据集的样本数量
        """
        return len(self.x)  # 数据集的样本数，x的长度即为数据集的大小
    
    def __getitem__(self, idx):
        """
        根据索引返回一个样本和其对应的标签。
        
        Args:
            idx (int): 索引
        
        Returns:
            tuple: 一个元组 (x, y)，x为输入特征，y为对应标签
        """
        # 返回给定索引 idx 对应的特征 x 和标签 y
        return self.x[idx], self.y[idx]
    
# 实例化训练集、验证集和测试集
# 这些实例化的对象可以通过DataLoader进行批量加载
train_ds = HousingDataset("train")  # 加载训练集数据
valid_ds = HousingDataset("valid")  # 加载验证集数据
test_ds = HousingDataset("test")  # 加载测试集数据
```

### DataLoader

```python
from torch.utils.data import DataLoader

#放到DataLoader中的train_ds, valid_ds, test_ds都是dataset类型的数据
batch_size = 16 #batch_size是可以调的超参数
train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(valid_ds, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_ds, batch_size=batch_size, shuffle=False)
```

## 定义模型

```python
#回归模型我们只需要1个数

class NeuralNetwork(nn.Module):
    def __init__(self, input_dim=8):
        super().__init__()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(input_dim, 30),
            nn.ReLU(),
            nn.Linear(30, 1)
            )
        
    def forward(self, x):
        # x.shape [batch size, 8]
        logits = self.linear_relu_stack(x)
        # logits.shape [batch size, 1]
        return logits
```

### 防止过拟合

```python
class EarlyStopCallback:
    def __init__(self, patience=5, min_delta=0.01):
        """

        Args:
            patience (int, optional): Number of epochs with no improvement after which training will be stopped.. Defaults to 5.
            min_delta (float, optional): Minimum change in the monitored quantity to qualify as an improvement, i.e. an absolute 
                change of less than min_delta, will count as no improvement. Defaults to 0.01.
        """
        self.patience = patience
        self.min_delta = min_delta
        self.best_metric = -1
        self.counter = 0
        
    def __call__(self, metric):
        if metric >= self.best_metric + self.min_delta:
            # update best metric
            self.best_metric = metric
            # reset counter 
            self.counter = 0
        else: 
            self.counter += 1
            
    @property
    def early_stop(self):
        return self.counter >= self.patience
```

```python
from sklearn.metrics import accuracy_score  # 导入用于计算准确率的sklearn库

# 定义评估函数，@torch.no_grad() 是一个装饰器，用于在评估时禁用梯度计算
@torch.no_grad()  
def evaluating(model, dataloader, loss_fct):
    """
    该函数用于在验证集上评估模型的表现，包括计算损失值（loss）。

    参数：
        model (nn.Module): 训练好的模型
        dataloader (DataLoader): 验证数据的dataloader，用于批量加载验证数据
        loss_fct (function): 损失函数，用于计算模型输出与实际标签之间的误差

    返回：
        float: 验证集的平均损失值
    """
    
    loss_list = []  # 用来存储每个批次的损失

    # 遍历验证集中的每个batch
    for datas, labels in dataloader:
        # 将数据和标签迁移到指定的设备（GPU或CPU）
        datas = datas.to(device)
        labels = labels.to(device)

        # 前向传播：使用训练好的模型预测每个批次的数据
        logits = model(datas)  # 模型输出的logits，通常是未经过激活函数的结果

        # 计算损失：将模型输出(logits)与标签(labels)输入到损失函数
        loss = loss_fct(logits, labels)  # 损失函数根据预测和真实标签计算误差

        # 将每个批次的损失值添加到loss_list中
        loss_list.append(loss.item())  # 使用 .item() 从tensor中提取出标量值，避免返回tensor对象

    # 计算并返回平均损失
    return np.mean(loss_list)  # 返回损失值的均值作为评估结果
```

```python
# 训练
def training(
    model, 
    train_loader, 
    val_loader, 
    epoch, 
    loss_fct, 
    optimizer,
    early_stop_callback=None,
    eval_step=500,
    ):
    record_dict = {
        "train": [],
        "val": []
    }
    
    global_step = 0
    model.train()
    with tqdm(total=epoch * len(train_loader)) as pbar:
        for epoch_id in range(epoch):
            # training
            for datas, labels in train_loader:#11610/16=741
                datas = datas.to(device)
                labels = labels.to(device)
                # 梯度清空
                optimizer.zero_grad()
                # 模型前向计算
                logits = model(datas) #得到预测值
                # 计算损失
                loss = loss_fct(logits, labels)
                # 梯度回传
                loss.backward()
                # 调整优化器，包括学习率的变动等
                optimizer.step()
 
                loss = loss.cpu().item() #转为cpu类型，item()是取值
                # record
                
                record_dict["train"].append({
                    "loss": loss, "step": global_step
                })
                
                # evaluating
                if global_step % eval_step == 0:
                    model.eval()
                    val_loss = evaluating(model, val_loader, loss_fct)
                    record_dict["val"].append({
                        "loss": val_loss, "step": global_step
                    })
                    model.train()

                    # 早停 Early Stop
                    if early_stop_callback is not None:
                        early_stop_callback(-val_loss) #根据验证集的损失来实现早停
                        if early_stop_callback.early_stop:
                            print(f"Early stop at epoch {epoch_id} / global_step {global_step}")
                            return record_dict
                    
                # udate step
                global_step += 1
                pbar.update(1)
                pbar.set_postfix({"epoch": epoch_id})
        
    return record_dict
        

epoch = 100

model = NeuralNetwork()

# 1. 定义损失函数 采用MSE损失,均方误差
loss_fct = nn.MSELoss()
# 2. 定义优化器 采用SGD
# Optimizers specified in the torch.optim package
optimizer = torch.optim.SGD(model.parameters(), lr=0.001, momentum=0.9)

# 3. early stop
early_stop_callback = EarlyStopCallback(patience=10, min_delta=1e-3)

model = model.to(device)
record = training(
    model, 
    train_loader, 
    val_loader, 
    epoch, 
    loss_fct, 
    optimizer, 
    early_stop_callback=early_stop_callback,
    eval_step=len(train_loader)
    )
```

```python
#画线要注意的是损失是不一定在零到1之间的
def plot_learning_curves(record_dict, sample_step=500):
    # build DataFrame
    train_df = pd.DataFrame(record_dict["train"]).set_index("step").iloc[::sample_step]
    val_df = pd.DataFrame(record_dict["val"]).set_index("step")

    # plot
    for idx, item in enumerate(train_df.columns):
        plt.plot(train_df.index, train_df[item], label=f"train_{item}")
        plt.plot(val_df.index, val_df[item], label=f"val_{item}")
        plt.grid()
        plt.legend()
        # plt.xticks(range(0, train_df.index[-1], 10*sample_step), range(0, train_df.index[-1], 10*sample_step))
        plt.xlabel("step")

        plt.show()

plot_learning_curves(record)  #横坐标是 steps
```
